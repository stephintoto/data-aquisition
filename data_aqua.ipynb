df.isnull().sum() calculates the number of missing values (NaN) in each column
missing_values[missing_values > 0] filters out columns that have no missing values, showing only those with missing data.

```python
missing_values = df.isnull().sum()
columns_missing_values = missing_values[missing_values > 0]
print(columns_missing_values)
```

df[categorical_column].unique() returns an array of the unique values in the specified column
df[categorical_column].nunique() returns the number of unique values in the column

```python
categorical_column = 'type'
unique_values = df[categorical_column].unique()
unique_count = df[categorical_column].nunique()
print(f"Unique values in '{categorical_column}': {unique_values}")
print(f"Number of unique values: {unique_count}")
```

df['duration'].mean() calculates the average duration
df['duration'].median() calculates the middle value
df['duration'].std() calculates the standard deviatio

```python
numerical_column = 'duration'
mean_value = df[numerical_column].mean()
median_value = df[numerical_column].median()
std_deviation = df[numerical_column].std()

# Print the results
print(f"Mean of '{numerical_column}': {mean_value}")
print(f"Median of '{numerical_column}': {median_value}")
print(f"Standard Deviation of '{numerical_column}': {std_deviation}")
```

Q1 and Q3: These are the 25th and 75th percentiles, represents the lower and upper bounds of the middle 50% of the data,
IQR: The range within which the central 50% of the data lies,
Lower and Upper Bounds: Any values below or above these bounds are considered outliers.

```python

num_col = 'duration'
Q1 = df[num_col].quantile(0.25)
Q3 = df[num_col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = df[(df[num_col] < lower_bound) | (df[num_col] > upper_bound)]

print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
print(f"Number of outliers in '{num_col}': {outliers.shape[0]}")
print(outliers[[num_col]])
```

.corr() method to find the correlation coefficient between the two numerical columns.
Positive Correlation: If the correlation is close to 1, it suggests that newer shows/movies tend to have longer or short durations.
Negative Correlation: If the correlation is close to -1, it suggests that older shows/movies tend to have longer or shorter duration.
No Correlation: If the correlation is close to 0, it suggests that there is no linear relationship between the release year and duration.

here it is close to 0 so there is no realation

```python

correlation = df['release_year'].corr(df['duration'])

print(f"Correlation between 'release_year' and 'duration': {correlation}")
```

groupby() method to group the DataFrame by the country column.

.agg() method to calculate various statistics like mean, sum, min, max, count for the duration column.

shows the agg stat of each country

```python

aggregated_stats = df.groupby('country')['duration'].agg(['mean', 'sum', 'min', 'max', 'count', 'std', 'median'])

print(aggregated_stats)
```
